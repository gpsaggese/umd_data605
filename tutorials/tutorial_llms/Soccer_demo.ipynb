```python
def parse_to_dict(data_str):
    # Split the input string by lines
    lines = data_str.strip().split('\n')
    
    # Initialize an empty dictionary
    data_dict = {}
    
    # Loop through each line
    for line in lines:
        if line == "":
            continue
        # Split each line by '=' to get key and value
        key, value = line.split('=', 1)
        # Strip any extra whitespace from key and value
        key = key.strip()
        value = value.strip()
        # Add the key-value pair to the dictionary
        data_dict[key] = value
    
    return data_dict

explan_dict = parse_to_dict(explanation)
```

```python
def _set_uniqueness_constraints(tx, node):
    query = f"""CREATE CONSTRAINT IF NOT EXISTS FOR (n:{node})
        REQUIRE n.id IS UNIQUE;"""
    _ = tx.run(query, {})


NODES = []
with driver.session(database="neo4j") as session:
    for node in NODES:
        session.execute_write(_set_uniqueness_constraints, node)
```

```python
// Create League and Division nodes
CREATE (league:League {name: 'Premier League'})
CREATE (division:Division {name: 'First Division'})

// Create Team nodes
CREATE (team1:Team {name: 'Team A'})
CREATE (team2:Team {name: 'Team B'})

// Create Referee node
CREATE (referee:Referee {name: 'Referee X'})

// Create Match node with details
CREATE (match:Match {date: '30/07/24', time: '15:00', full_time_result: 'H', half_time_result: 'D'})

// Create Match Detail nodes for full time and half time
CREATE (fullTime:MatchDetail {type: 'full_time', home_goals: 2, away_goals: 1, result: 'H'})
CREATE (halfTime:MatchDetail {type: 'half_time', home_goals: 1, away_goals: 1, result: 'D'})

// Create Statistics nodes for both teams
CREATE (homeStats:Statistic {shots: 10, shots_on_target: 6, fouls_committed: 12, corners: 5, yellow_cards: 1, red_cards: 0})
CREATE (awayStats:Statistic {shots: 8, shots_on_target: 4, fouls_committed: 14, corners: 3, yellow_cards: 2, red_cards: 1})

// Create relationships
CREATE (league)-[:CONTAINS]->(division)
CREATE (division)-[:INCLUDES]->(match)
CREATE (match)-[:HOME_TEAM]->(team1)
CREATE (match)-[:AWAY_TEAM]->(team2)
CREATE (match)-[:REFEREED_BY]->(referee)
CREATE (match)-[:HAS_DETAIL]->(fullTime)
CREATE (match)-[:HAS_DETAIL]->(halfTime)
CREATE (match)-[:HAS_STATISTICS {type: 'home'}]->(homeStats)
CREATE (match)-[:HAS_STATISTICS {type: 'away'}]->(awayStats)
```
