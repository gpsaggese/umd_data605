```python
def parse_to_dict(data_str):
    # Split the input string by lines
    lines = data_str.strip().split('\n')
    
    # Initialize an empty dictionary
    data_dict = {}
    
    # Loop through each line
    for line in lines:
        if line == "":
            continue
        # Split each line by '=' to get key and value
        key, value = line.split('=', 1)
        # Strip any extra whitespace from key and value
        key = key.strip()
        value = value.strip()
        # Add the key-value pair to the dictionary
        data_dict[key] = value
    
    return data_dict

explan_dict = parse_to_dict(explanation)
```

```python
def _set_uniqueness_constraints(tx, node):
    query = f"""CREATE CONSTRAINT IF NOT EXISTS FOR (n:{node})
        REQUIRE n.id IS UNIQUE;"""
    _ = tx.run(query, {})


NODES = []
with driver.session(database="neo4j") as session:
    for node in NODES:
        session.execute_write(_set_uniqueness_constraints, node)
```

```python
# Define a function to create a match node and relationships
def create_match(tx, match_data):
    query = """
    MERGE (league:League {name: $league})
    MERGE (division:Division {name: $division})
    MERGE (division)-[:BELONGS_TO]->(league)
    MERGE (homeTeam:Team {name: $homeTeam})
    MERGE (awayTeam:Team {name: $awayTeam})
    MERGE (referee:Referee {name: $referee})
    MERGE (match:Match {id: $matchId})
    ON CREATE SET
      match.date = $date,
      match.time = $time,
      match.fullTimeHomeGoals = $fullTimeHomeGoals,
      match.fullTimeAwayGoals = $fullTimeAwayGoals,
      match.fullTimeResult = $fullTimeResult,
      match.halfTimeHomeGoals = $halfTimeHomeGoals,
      match.halfTimeAwayGoals = $halfTimeAwayGoals,
      match.halfTimeResult = $halfTimeResult,
      match.homeTeamShots = $homeTeamShots,
      match.awayTeamShots = $awayTeamShots,
      match.homeTeamShotsOnTarget = $homeTeamShotsOnTarget,
      match.awayTeamShotsOnTarget = $awayTeamShotsOnTarget,
      match.homeTeamFoulsCommitted = $homeTeamFoulsCommitted,
      match.awayTeamFoulsCommitted = $awayTeamFoulsCommitted,
      match.homeTeamCorners = $homeTeamCorners,
      match.awayTeamCorners = $awayTeamCorners,
      match.homeTeamYellowCards = $homeTeamYellowCards,
      match.awayTeamYellowCards = $awayTeamYellowCards,
      match.homeTeamRedCards = $homeTeamRedCards,
      match.awayTeamRedCards = $awayTeamRedCards
    MERGE (match)-[:PLAYED_IN]->(division)
    MERGE (match)-[:HOME_TEAM]->(homeTeam)
    MERGE (match)-[:AWAY_TEAM]->(awayTeam)
    MERGE (match)-[:REFEREED_BY]->(referee)
    """
    tx.run(query, match_data)

# Function to import data from DataFrame to Neo4j
def import_data_from_dataframe(df, driver):
    for index, row in df.iterrows():
        match_data = {
            "league": row["League Division"].split()[0],
            "division": row["League Division"],
            "date": row["Match Date"],
            "time": row["Time of match kick-off"],
            "homeTeam": row["Home Team"],
            "awayTeam": row["Away Team"],
            "fullTimeHomeGoals": row["Full Time Home Team Goals"],
            "fullTimeAwayGoals": row["Full Time Away Team Goals"],
            "fullTimeResult": row["Full Time Result"],
            "halfTimeHomeGoals": row["Half Time Home Team Goals"],
            "halfTimeAwayGoals": row["Half Time Away Team Goals"],
            "halfTimeResult": row["Half Time Result"],
            "referee": row["Match Referee"],
            "homeTeamShots": row["Home Team Shots"],
            "awayTeamShots": row["Away Team Shots"],
            "homeTeamShotsOnTarget": row["Home Team Shots on Target"],
            "awayTeamShotsOnTarget": row["Away Team Shots on Target"],
            "homeTeamFoulsCommitted": row["Home Team Fouls Committed"],
            "awayTeamFoulsCommitted": row["Away Team Fouls Committed"],
            "homeTeamCorners": row["Home Team Corners"],
            "awayTeamCorners": row["Away Team Corners"],
            "homeTeamYellowCards": row["Home Team Yellow Cards"],
            "awayTeamYellowCards": row["Away Team Yellow Cards"],
            "homeTeamRedCards": row["Home Team Red Cards"],
            "awayTeamRedCards": row["Away Team Red Cards"],
            "matchId": row["Match Date"] + row["Time of match kick-off"] + row["Home Team"] + row["Away Team"]  # Unique identifier
        }
        with driver.session() as session:
            session.write_transaction(create_match, match_data)

import_data_from_dataframe(df2, driver)
```
